```python
from kafka import KafkaConsumer
import json

# Setup Consumer
consumer = KafkaConsumer(
    'transactions',
    bootstrap_servers=['localhost:9092'],
    auto_offset_reset='earliest',
    group_id='json-writer-group',
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))
)

print("Reading from Kafka... Press Ctrl+C to stop.")

with open('output_transactions.json', 'a') as f:
    try:
        for message in consumer:
            transaction = message.value
            # Write each record as a new line in the JSON file
            f.write(json.dumps(transaction) + "\n")
            f.flush() # Ensure it writes to disk immediately
            print(f"Saved: {transaction['txn_id']}")
    except KeyboardInterrupt:
        print("Stopping consumer...")
```

```python
import pandas as pd

# Load the JSON file
# If your JSON is a list of dictionaries, pandas reads it directly
df = pd.read_json('output_transactions.json',lines = True)

# Save to CSV
df.to_csv('output_transactions.csv', index=False)

print("Conversion complete: data.csv created.")
```
